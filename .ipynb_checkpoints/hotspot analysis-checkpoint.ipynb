### Load data

```r
library(ggplot2)
library(sqldf)
options(encoding="UTF-8")
```

Get the daily hotspot count within 50 kilometers around Nan. Create date type from acq_date column
The daily hotspot is between 22-02-2018 and 30-04-2018

```r
hotspot<-read.csv("https://raw.githubusercontent.com/Maplub/AirQualityData/master/count_50km.csv")
colnames(hotspot)<-c("id","acq_date","count")
hotspot$date<-as.POSIXct(strptime(hotspot$acq_date, "%Y-%m-%d"))
```

```r
head(hotspot)
```

```r
min(hotspot$date)
max(hotspot$date)
```

Get PM2.5 Daily Data

```r
daily<-read.csv("https://raw.githubusercontent.com/Maplub/AirQualityData/master/PM2-5_2018_daily.csv")
colnames(daily)<-c('sensorid','date','pm25')
head(daily)
```

```r
daily$timepos <- as.POSIXct(strptime(daily$date, "%Y-%m-%d"))
```

Select PM2.5 data between 22 Feb and 30 Apr 2018

```r
newdaily<-daily[daily$timepos>="2018-02-22" & daily$timepos<="2018-04-30",]
newdaily<- newdaily[order(newdaily$timepos, newdaily$sensorid),]
```

### Deal with missing values

Create full date daily with distinct sensor id from newhour

```r
datefull<-seq(from=min(newdaily$timepos),to=max(newdaily$timepos), by="day")
```

```r
sensor_newdaily<-sqldf('select distinct(sensorid) from newdaily')
```

merge data to get every sensor with full datetime


```r
fulldata<-merge(sensor_newdaily,datefull)
colnames(fulldata)[2]<-'timepos'
```

merge full data with raw data. leave null values จะได้ข้อมูลที่เว้นว่างวันที่ไม่มีข้อมูล

```r
newdaily<-sqldf('select f.sensorid,f.timepos,n.pm25 from fulldata f LEFT JOIN newdaily n ON f.sensorid = n.sensorid AND f.timepos=n.timepos')
```

```r
count_na<-aggregate(pm25 ~ sensorid, data=newdaily, function(x) {sum(is.na(x))}, na.action = NULL)
sensorFull<-count_na$sensorid[which(count_na$pm25<10)]
sensorFull
```

```r
newdaily<-subset(newdaily,sensorid %in% sensorFull)
```

```r
newdaily$dow<-as.numeric(format(newdaily$timepos,"%w"))
```

Plot PM2.5 data to see if there is an outlier

```r
ggplot(data=newdaily, aes(x=timepos, y=pm25, group=sensorid)) + geom_line()
```

```r
sapply(newdaily, class)
```

Remove extreme value and plot the data by day of week

```r
remove_extreme<-sqldf("select * from newdaily where pm25<2500")
ggplot(data=remove_extreme, aes(x=factor(dow), y=pm25,group=factor(dow))) + geom_boxplot()+ scale_x_discrete(name="Day of Week",labels=c("Sunday", "Monday", "Tuesday","Wednesday","Thursday","Friday","Saturday"))
```

```r
avg_daily_pm25<-sqldf("select timepos,avg(pm25) as avgpm25 from newdaily group by timepos")
```

```r
datadaily<-sqldf('select a.timepos,a.avgpm25,h.count from avg_daily_pm25 a LEFT JOIN hotspot h ON a.timepos=h.date')
```

Plot a time series of PM2.5 (in blue) and a hotspot count (in red). Normalized the values to 0-1 before plotting. Similar trend can be seen. As the number of count rises, PM2.5 also rises.

```r
range01 <- function(x){(x-min(x))/(max(x)-min(x))}
datadaily$count[is.na(datadaily$count)] <- 0

ggplot(data=datadaily,aes(x=timepos)) + geom_line(aes(y=range01(avgpm25)), color='blue') + geom_line(aes(y=range01(count)), color='red')+ labs(y = "Normalized PM2.5")
```

Plot the PM2.5 VS Count. There is a correlation between them.

```r
ggplot(data=datadaily,aes(x=avgpm25,y=count)) + geom_point() + geom_smooth()  
```

## Now I will analyze time lag between PM2.5 and hotspot count. When the hotspot rises, will PM2.5 value rises at the same or there is a time lag between them?

Try CCF plot between PM2.5 and count. The result shows lag 1 and cyclical pattern

```r
library(xts)
t1<-xts(datadaily$avgpm25,datadaily$timepos)
t2<-xts(datadaily$count,datadaily$timepos)
ccf(as.numeric(t1), as.numeric(t2),na.action = na.pass) 
```

## We plot the ACF separately to see the detail of each data.

### ACF of average PM2.5

```r
acf(t1,na.action=na.pass)
```

### ACF of hotspot counts

```r
acf(t2,na.action=na.pass)
```

### Some said that hotspot shows day of week pattern. I will plot day of week heat map to see if there is a pattern.

```r
head(datadaily)
```

```r
datadaily$dow<-as.POSIXlt(datadaily$timepos)$wday
```

```r
datadaily$dowf<-factor(datadaily$dow,levels=rev(0:6),labels=rev(c("Sun","Mon","Tue","Wed","Thu","Fri","Sat")),ordered=TRUE)
```

```r
library(lubridate)
datadaily$d<-day(datadaily$timepos)
datadaily$m<-month(datadaily$timepos)
datadaily$y<-year(datadaily$timepos)
```

```r
datadaily$woy<-as.numeric(format(datadaily$timepos,"%W"))
library(plyr)
datadaily<-ddply(datadaily,.(m),transform,monthweek=1+woy-min(woy))
```

```r
sapply(datadaily, class)
```

```r
ggplot(datadaily, aes(monthweek,dowf)) + geom_tile(aes(fill = count))+  scale_fill_continuous(high = "#132B43", low = "#56B1F7") + facet_grid(y~m) + xlab("Week of Month") + ylab("")
```

```r
ggplot(datadaily, aes(monthweek,dowf)) + geom_tile(aes(fill = avgpm25))+  scale_fill_continuous(high = "#132B43", low = "#56B1F7") + facet_grid(y~m) + xlab("Week of Month") + ylab("")
```
