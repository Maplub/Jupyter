```r
pcd$y<-paste("20",substr(pcd[,'date'], 1, 2),sep="")
pcd$m<-substr(pcd[,'date'], 3, 4)
pcd$d<-substr(pcd[,'date'], 5, 6)
```

```r
pcd$date<-paste(pcd$y,pcd$m,pcd$d,sep="-")
```

```r
pcd$date<-as.Date(format(as.Date(pcd$date), '%Y-%m-%d'))
```

```r
data<-sqldf("select p.date,p.y,p.m,p.d,p.hour,p.pm25 as pm25, p.pm10 as pm10, c.avgpm25 as mypm25, c.avgpm10 as mypm10 
from pcd p LEFT JOIN chula c ON p.date=c.date and p.hour=c.h*100 order by p.date")
```

```r
data$pm25<-as.numeric(data$pm25)
data$pm10<-as.numeric(data$pm10)
```

```r
data$mypm25<-as.numeric(data$mypm25)
data$mypm10<-as.numeric(data$mypm10)
```

```r
data$numrow<-c(1:nrow(data))
```

```r
data$h<-data$hour/100
```

```r
sub1<-sqldf("select * from data where m='03'")
```

```r
library(ggplot2)
```

```r
rowstart<-sqldf('select numrow,date from sub1 where h=1')
rowstartdate<-rowstart[,1]
datelabel<-rowstart[,2]
```

```r
ggplot(sub1, aes(numrow)) + scale_x_continuous(name ="date", breaks = rowstartdate,label = datelabel) + 
geom_line(aes(y = pm25, colour = "PM25_PCD")) + geom_line(aes(y = mypm25, colour = "PM25_chula"))+
theme(text = element_text(size=8),axis.text.x = element_text(angle=90, hjust=1)) 
```
